## 1. Graficar las siguientes funciones f(x). Determina el dominio a partir de la función:



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 6, 300)  
y = x**2 - 4*x + 2

# Graficar la función
plt.figure(figsize=(8, 6))
plt.plot(x, y,label='"f(x) = x^2 - 4x + 2"' )
plt.title(" función f(x) = x^2 - 4x + 2")
plt.xlabel("x")
plt.ylabel("y=f(x)")
plt.grid()
plt.legend() # este activa la funcion en el grafico 

```

EL dominio de la funcion son todos los numeros reales va de (-∞ ,∞) ya que es un polinomio.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 6, 300)  
y = np.sin(3*x)

# Graficar la función
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="sin(3*x)")
plt.title(" función f(x) =sin(3x)")
plt.xlabel("x")
plt.ylabel("y=f(x)")
plt.grid()
plt.legend()
```

Dominio de sin(3x) son todos los numeros reales.
es el mismo que el dominio de la función seno básica, ya que la multiplicación 
del argumento por una constante no cambia el conjunto de valores para los cuales la función está definida

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return np.sqrt(x**2 - 4)

x1 = np.linspace(-4, -2, 300)  
x2 = np.linspace(2, 4, 300)    

y1 =f(x1)
y2 =f(x2)

# Graficar la función
plt.figure(figsize=(8, 6))
plt.plot(x1, y1, label=r"f(x) =√x^2-4 ", color='green')
plt.plot(x2, y2, color='green')
plt.title(" función f(x) = √x^2-4")
plt.xlabel("x")
plt.ylabel("y=f(x)")
plt.grid()
plt.legend()
```

El dominio es: (−∞,−2]∪[2,∞), ya que la grafica se ve que termina

## 3. Crea una función que cree una gráfica y dibuje 'k' líneas

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_sin_lines(x1, x2, n, a, b, k):
    
    
    x = np.linspace(x1, x2, n)
    valores_alpha = np.linspace(a, b, k)
    
    # Crear la gráfica
    plt.figure(figsize=(10, 6))
     # Graficar cada línea
    for alpha in valores_alpha:
        y = np.sin(alpha * x)
        plt.plot(x, y, label=f'α = {alpha:.2f}')  # Etiqueta con valor de α
        
    plt.title('Gráfico de f(x) = sin(αx) ')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid()
    plt.show()

# Ejemplo de uso
plot_sin_lines(x1=0, x2=2, n=300, a=2, b=4, k=3)
```
