## 1. Graficar las siguientes funciones f(x). Determina el dominio a partir de la funci√≥n:



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 6, 300)  
y = x**2 - 4*x + 2

# Graficar la funci√≥n
plt.figure(figsize=(8, 6))
plt.plot(x, y,label="$f(x) = x^2 - 4x + 2$" )
plt.title(" $funci√≥n f(x) =x^2 - 4x + 2$")
plt.xlabel("x")
plt.ylabel("y=f(x)")
plt.grid()
plt.legend() # este activa la funcion en el grafico 

```

EL dominio de la funcion son todos los numeros reales va de (-‚àû ,‚àû) ya que es un polinomio.

## 3. Crea una funci√≥n que cree una gr√°fica y dibuje 'k' l√≠neas

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_sin_lines(x1, x2, n, a, b, k):
    
    
    x = np.linspace(x1, x2, n)
    valores_alpha = np.linspace(a, b, k)
    
    # Crear la gr√°fica
    plt.figure(figsize=(10, 6))
     # Graficar cada l√≠nea
    for alpha in valores_alpha:
        y = np.sin(alpha * x)
        plt.plot(x, y, label=f'Œ± = {alpha:.2f}')  # Etiqueta con valor de Œ±
        
    plt.title('Gr√°fico de f(x) = sin(Œ±x) ')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid()
    plt.show()

# Ejemplo de uso
plot_sin_lines(x1=0, x2=2, n=300, a=2, b=4, k=3)
```
